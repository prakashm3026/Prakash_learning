```python
#importing the library for handling tables and dataset
import pandas as pd
```

```python
#loading the data set in to the variable dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#displaying the dataset
dataset
```

```python
#assiging the input to the variable independent --> iNPUT/OUTPUT SPLIT
independent=dataset[["YearsExperience"]]
```

```python
#displaying the input variable
independent
```

```python
#Assiging the outout column to the variable dependent --> INPUT/OUTPUT SPLIT
dependent=dataset[["Salary"]]
```

```python
#Displaying the salary column
dependent
```

```python
# splitting the dataset to test and train sets using sklearn -->SPLIT TRAIN TEST
from sklearn.model_selection import train_test_split
train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# assiginng train and test variable accordingly 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Creating SLR model using sklearn library
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#assigning the weight value to variable weight
weight=regressor.coef_
weight
```

```python
#Assiging the bias to the vairable
bias=regressor.intercept_
bias
```

```python
#finding the prediction value using test input
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
#finding the r_score value to find the correctness of the prediciton(Evaluation method)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#displaying rscore value
r_score
```

```python
#Saving the model in to file--> creating file name
import pickle 
filename="final_model_simple_linear.sav"
```

```python
#Saving the model in to file--> saving the model in to file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#executing the saved model
load_model=pickle.load(open("final_model_simple_linear.sav",'rb'))
results=load_model.predict([[20]])
results
```
