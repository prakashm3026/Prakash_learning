```python
import pandas as pd
```

```python
#read the csv file
dataset=pd.read_csv("C:\Prakash\AI\Learning\Machine_learning\Decision_tree\\50_startup_profit.csv")
```

```python
dataset
```

```python
#removing the dummy data or duplicate data

dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#displaying the headers of the columns
dataset.columns
```

```python
#assiging the input to the variable independent 
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
#assiging the output to the variable dependent 
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
# assiginng train and test variable accordingly 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor( criterion='friedman_mse',splitter='best',max_features='sqrt')
regressor=regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree 
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
r_score
```

```python
#Saving the model in to file--> creating file name

import pickle 
filename="final_model_dt.sav"
```

```python
#Saving the model in to file--> saving the model in to file
pickle.dump(regressor,open(filename,'wb'))
```

```python
#executing the saved model
load_model=pickle.load(open("final_model_dt.sav",'rb'))
results=load_model.predict([[12345,345,4565,1,0]])
results
```
